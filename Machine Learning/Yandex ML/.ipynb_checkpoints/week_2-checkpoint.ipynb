```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
import math
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
# True for features normalization
is_norm_features = True

col_names = ['Type', 'Alcohol ', 'Malic acid ', 'Ash ', 'Alcalinity of ash ', 'Magnesium ',
'Total phenols ', 'Flavanoids ', 'Nonflavanoid phenols ', 'Proanthocyanins ', 'Color intensity ',
'Hue ', 'OD280/OD315', 'Proline ']

# Load data
data = pd.read_csv('wine.csv', header = None)
data.columns = col_names

# Create class and features vectors
df_class = pd.DataFrame(data['Type'].copy(), columns=['Type'])
df_features = data.copy()
del df_features['Type']

# Normalize featues if nesessary
if is_norm_features:
    df_features = sklearn.preprocessing.scale(df_features)

# Create cross validation groupes for 5 splits
kf = KFold(n_splits=5, shuffle=True, random_state=42)

size = 50
mean_scores = []
for cur_neigh_numb in range(1, size+1):
    # Create classifier - k-means
    neigh =  KNeighborsClassifier(cur_neigh_numb)
    # Obtain cross validation scores for all 5 splits
    scores = cross_val_score(neigh, df_features, df_class, cv=kf)
    # Get mean score by all five splits
    mean_scores.append(np.mean(scores))

max_id = np.argmax(mean_scores)
print('Optimal k  = {0}, Score = {1}'.format(max_id, mean_scores[max_id]))

with open('2_1.txt', 'w') as out_f:
    out_f.write(str(max_id + 1))

with open('2_2.txt', 'w') as out_f:
    ans = '{0:.2f} '.format(mean_scores[max_id])
    out_f.write(ans)
```

Below I plot find out optimal number of neighbors for iris dataset in 2D space, but for 3 classes using LeaveOneOut cross validation.

```python
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score

from sklearn import neighbors
from sklearn.model_selection import LeaveOneOut, LeavePOut
from sklearn.model_selection import cross_val_score

iris = load_iris()
X = iris.data[:,0:2]
y = iris.target

# Find optimal number of neighbors
loo = LeaveOneOut()
n = float(loo.get_n_splits(X))
neighbors_numb = 100
err = np.zeros(neighbors_numb-1)

for k in range(1, neighbors_numb):
    clf = neighbors.KNeighborsClassifier(n_neighbors=k, weights='uniform')
    err[k-1] = 1.-np.mean(cross_val_score(clf, X, y, cv=loo, scoring='accuracy'))

opt_numb = np.argmin(err) + 1
print('Optimal neighbors number is {0}'.format(opt_numb))
```

- Dependency of error from number of neighbors in calssification model. We can find out that **k=48** is the best. Here we use **LOO cross-validation to obtain optimal number of neighbors k**.
- Decision boundary for k=1. Here we can see that with only one neighbor classification algorithm show bad boundary, because it reacts on **WRONG EMISSIONS**, so we can find red area close to left nearles corner, but it should not be there !!!! This is an green point emission. And near we can see some several emissions of yellow points inside the green area.
- Decision boundary for k=48. Shows us quite good decision boundary, which consists from 3 well separated areas **WHITHOUT EMISSIONS**. Good result.

Below I will illustrate **comparation of two algorithms**: kNN and Parzen window (Both are implemented in one class, because I suck).

- In kNN we need to choose **number of neighbors**, and calculate predicted value based onlly on this neighbors, while weights could be equal to 1, or in some other way.
- In Parzen window we need to calcualte predicted value using ** all train dataset**, but **weights calculation is based on kernel type (below is Gauss kernel) and width of Parzen window (below h parameter)**.

As a result we can see that **kNN** with 1 neighbor is equivalent to **Parzen window classifier** with h=0.1 (Little window width), and kNN(n_neighbors=6) is equal Parzen(h=0.24).
So: **this two methods are aquivalent with appropriate parameters choise (number of neithbors/window width)**.

```python
from sklearn.datasets import make_circles

class ParzenKNN:
    # Iniitalize:
    # 1. For kNN classifier -> number of neighbors
    # 2. For Parzen window classifier -> h
    def __init__(self, neigh_num, parzen_h):
        self.n_neighbors = neigh_num
        self.parzen_h = parzen_h
    
    # Store train dataset
    def fit(self, X, y):
        self.X = X
        self.y = y
    
    # Cacluclate Euclidian distance between X_train and 
    # single element in X_test
    def distance(self, X_test):
        r = np.zeros(self.X.shape[0])
        for i in range(r.shape[0]):
            r[i] = np.sqrt(np.sum((self.X[i] - X_test)**2))
        return r   
    
    # Predictions for kNN classifier
    def predict_simple_point_knn(self, X_test):
        # Calculate distances from test obj. to objects in train dataset
        r = self.distance(X_test) / self.parzen_h
        # Choose ids of 'n_neighbors' closest to test obj objects form train dataset
        closest_idx = np.argpartition(r, self.n_neighbors)[:self.n_neighbors]
        # Calculate parzen distances
        r = r[closest_idx] / self.parzen_h
        # Use Gausian kernel for weights calculation
        weights = np.ones(r.shape[0])#np.exp(- 2.* r**2)
        # Calculate predicted value
        predict = np.sum(weights*self.y[closest_idx])
        if predict > 0:
            return 1
        else:
            return -1
        
    # Predictions for Parzen window classifier
    def predict_simple_point_parzen(self, X_test):
        r = self.distance(X_test) / self.parzen_h
        weights = np.exp(- 2.* r**2)
        predict = np.sum(weights*self.y)
        if predict > 0:
            return 1
        else:
            return -1
    
    # Predict for all dataset with choosen classifier
    def predict(self, X_test, pred_type):
        pred = []
        if pred_type == 'parzen':
            for x_test in X_test:
                pred.append(self.predict_simple_point_parzen(x_test))
        elif pred_type == 'knn':
            for x_test in X_test:
                pred.append(self.predict_simple_point_knn(x_test))
        return np.array(pred)

if __name__ == '__main__':
    # Generate train dataset
    X, y = make_circles(n_samples=300, noise=0.2, random_state=241, factor=0.5)
    y[y == 0] = -1
    # Generate test dataset (for area draw)
    area = [50, 50]
    x_ = np.linspace(np.min(X[:,0])-0.2, np.max(X[:,0])+0.2, area[0] )
    y_ = np.linspace(np.min(X[:,1])-0.2, np.max(X[:,1])+0.2, area[1] )

    X_test = [[i, j] for i in x_ for j in y_]
    
    
    # Classification 1-neighbor and parzen with h=0.1
    clf = ParzenKNN(1, 0.1)
    clf.fit(X,y)
    y_test = (clf.predict(X_test, pred_type='parzen')).reshape((area[0], area[1]))
    
    fig, ax = plt.subplots(2, 2, figsize=(12,8))
    
    ax[0,0].contourf(x_,y_,y_test.T,  cmap=plt.cm.coolwarm, alpha=0.5)
    ax[0,0].scatter(X[:,0],X[:,1],c=y)
    ax[0,0].set_title('Classifier with Parzen window, h = 0.1', fontsize=15)
    
    y_test = (clf.predict(X_test, pred_type='knn')).reshape((area[0], area[1]))
    ax[0,1].contourf(x_,y_,y_test.T,  cmap=plt.cm.coolwarm, alpha=0.5)
    ax[0,1].scatter(X[:,0],X[:,1],c=y)
    ax[0,1].set_title('kNN Classifier with n_neighbors = 1.', fontsize=15)
    
    # Classification 5-neighbors and parzen with h=0.5
    clf = ParzenKNN(6, 0.24)
    clf.fit(X,y)
    y_test = (clf.predict(X_test, pred_type='parzen')).reshape((area[0], area[1]))
    
    ax[1,0].contourf(x_,y_,y_test.T,  cmap=plt.cm.coolwarm, alpha=0.5)
    ax[1,0].scatter(X[:,0],X[:,1],c=y)
    ax[1,0].set_title('Classifier with Parzen window, h = 0.24', fontsize=15)
    
    y_test = (clf.predict(X_test, pred_type='knn')).reshape((area[0], area[1]))
    ax[1,1].contourf(x_,y_,y_test.T,  cmap=plt.cm.coolwarm, alpha=0.5)
    ax[1,1].scatter(X[:,0],X[:,1],c=y)
    ax[1,1].set_title('kNN Classifier with n_neighbors = 6.', fontsize=15)
    
    plt.show()
```
