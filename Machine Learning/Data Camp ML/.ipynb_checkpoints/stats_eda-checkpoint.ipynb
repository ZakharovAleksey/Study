## EDA (exploratory data analysis)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import datasets

sns.set()
```

Import the iris dataset

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target

df = pd.DataFrame(data = np.c_[X, y], columns = iris['feature_names'] + ['target'])
df.head()
```

### BinSwarm Plot

is more usefull than histograms and more representative.

```python
fig, ax = plt.subplots(figsize=(10, 5))

_ = sns.swarmplot(
    x = 'target',
    y = 'sepal length (cm)',
    data = df,
    ax = ax
)

_ = plt.xlabel('target')
_ = plt.ylabel('sepal length (cm)')

plt.show()
```

### Plot ECDF with percentiles

```python
def ecdf(data):
    n = data.shape[0]
    x = np.sort(data)
    # Probabilities
    y = np.arange(1, n + 1, 1.) / n
    
    return x, y

fig, ax = plt.subplots(figsize=(10, 5))

# Calculate ECDF
x_plot, y_plot = ecdf(df['sepal length (cm)'])

# Calculate percentiles
perc = np.array([2.5, 25, 50, 75, 97.5])
perc_vals = np.percentile(x_plot, perc)

# Plot ECDF & percentiles
_ = plt.plot(x_plot, y_plot, marker = '.', linestyle = 'none', color = 'C1')
_ = plt.plot(perc_vals, perc / 100., marker = 'D', linestyle = 'none', color = 'C2')

_ = plt.xlabel('sepal length (cm)')
_ = plt.ylabel('Cummulative probability')

plt.show()
```

## Boxplot

```python
fig, ax = plt.subplots(figsize=(10, 5))

_ = sns.boxplot(
    x = 'target',
    y = 'sepal length (cm)',
    data = df,
    ax = ax
)

_ = plt.xlabel('target')
_ = plt.ylabel('sepal length (cm)')

plt.show()
```

## ScatterPlot

is used for correlation visual demonstration.

```python
fig, ax = plt.subplots(figsize=(10, 5))

_ = df.plot.scatter(
    x = 'petal length (cm)',
    y = 'sepal length (cm)',
    ax = ax
)

_ = plt.xlabel('petal length (cm)')
_ = plt.ylabel('sepal length (cm)')

print(
    'Correlation between two features is {:.3f}'.format(
        df[
            ['petal length (cm)', 'sepal length (cm)']
        ].corr().loc['petal length (cm)', 'sepal length (cm)']
    )
)


plt.show()
```

## Botstrap Replicates

Therminology: *botstrap* - re-simulated array of input data, while *replicate* is some statistics of re-simulated data.

This technique allows repeat dataset values: but why do we need theis twechnique?
- As far as I unredstood, we could not repeat an experiment infinite number of times in some cases. But using this technique **we could simulate repetitions on an experiment**.

Here we will use np.random.choise(data, len(data)) function to repeat the experiment.

**Conclusion**: this technique allows us to estimate accurasy of replicates, like variance, mean, etc. because we obtain their distribution (see pic. below) and we could estimate confidance interval for this values. 
And confidense interval tell us with what accuracy we could trust measured descriptive statistics.

```python
# Calculates array of replicates for chosen data
def generate_replicate(data, func, size = 1):
    replicates = np.empty(size)
    for i in range(size):
        replicates[i] = func(np.random.choice(data, size=len(data)))
    return replicates
```

```python
# Here we plot distribution of several replicates for data
data = np.random.normal(10, 2, 10000)
funcs = [np.mean, np.var, np.std]
labels = ['mean', 'var', 'std']

fig, ax = plt.subplots(figsize=(10, 5))

for index, cur_func in enumerate(funcs):
    cur_repl = generate_replicate(data, cur_func, 1000)
    print('{0} : {1}\t 95% confident interval = {2}'.format(
        labels[index], 
        round(np.mean(cur_repl), 2),
        map(lambda x: round(x, 2), np.percentile(cur_repl, [2.5, 97.5])))
    )

    ax.hist(cur_repl, bins = 50, alpha = 0.5, label = labels[index])

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
    
plt.show()
```

Note: we have a normal distribution for all descriptive statistics above (for mean, variance, standart deviation).


### Plus we could use botstrap in pair cases, like in lineal regression.

Here idea - is to select indexes op chosen pairs.

```python
def slope_inrtersect_pair_replicates(x, y, size=1):
    ids = np.arange(x.shape[0])
    
    bs_slope, bs_intersect = np.empty(size), np.empty(size)
    
    for i in range(size):
        bs_ids = np.random.choice(ids, size=len(ids))
        bs_x, bs_y = x[bs_ids], y[bs_ids]
        bs_slope[i], bs_intersect[i] = np.polyfit(bs_x, bs_y, 1)
    return bs_slope, bs_intersect
```

```python
np.random.seed(20)
x_measured = np.random.normal(10, 2, 100)
y_measured = np.random.normal(5, 1, 100)

x = np.array([np.min(x_measured), np.max(x_measured)])

fig, ax = plt.subplots(figsize = (10, 5))

ax.scatter(x_measured, y_measured, alpha = 0.5, color = 'C1')

# Boststrap lregression lines
bs_slope, bs_inter = slope_inrtersect_pair_replicates(x_measured, y_measured, 100)
for i in range(100):
    ax.plot(x, bs_slope[i] * x + bs_inter[i], color = 'C2', alpha = 0.2)

#  Main regression line
slope, intersect = np.polyfit(x_measured, y_measured, 1)
ax.plot(x, slope * x + intersect, marker='.', color = 'k', lw=3)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Dependency of regression lines on botstrap values')
plt.show()
```

# Permutation.

```python
def perf_permutation(data_1, data_2):
    
    data = np.concatenate((data_1, data_2))
    per_data = np.random.permutation(data)
    
    return per_data[:len(data_1)], per_data[len(data_1):]
```

```python
data_1 = np.random.normal(0, 2, size=200)
data_2 = np.random.normal(0.01, 1.9, size=200)
data_3 = np.random.standard_t(10,  size=200)

fig, ax = plt.subplots(figsize=(10, 5))

_ = ax.hist(data_1, bins= 50, alpha = 0.5, label = 'data_1')
_ = ax.hist(data_2, bins= 50, alpha = 0.5, label = 'data_2')
_ = ax.hist(data_3, bins= 50, alpha = 0.5, label = 'data_3')

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

for i in range(10):
    pr_1, pr_2 = perf_permutation(data_1, data_3)
    x_p1, y_p1 = ecdf(pr_1)
    x_p2, y_p2 = ecdf(pr_2)
    
    ax.plot(x_p1, y_p1, 
            color='C2', marker='.', linestyle='none', lw=2, alpha=0.1)
    ax.plot(x_p2, y_p2, 
            color='C3', marker='.', linestyle='none', lw=2, alpha=0.1)

x_1, y_1 = ecdf(data_1); 
x_2, y_2 = ecdf(data_3)

ax.plot(x_1, y_1, color='C2', marker='.', linestyle='none', lw=2)
ax.plot(x_2, y_2, color='C3', marker='.', linestyle='none', lw=2)

    
plt.show()
```
